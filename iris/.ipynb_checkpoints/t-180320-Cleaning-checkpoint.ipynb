```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_style('white')
sns.set_context("poster")
```

```python
col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df = pd.read_table('data-iris.txt', delimiter=',', names=col_names)
df.head()
```

```python
df.dtypes
```

```python
# clean class
df['class'] = df['class'].str.replace('Iris-', '')
```

```python
# set col "class" to category type - reduce memeory usage
df['class'] = df['class'].astype('category')
```

```python
# Summarise data for each class
df.describe()
```

```python
# Summarise data for each class
# seperate different classes to different df

classes = df['class'].unique()
dfs = []
for cls in classes:
    print(cls)
    print(df[df['class'] == cls].describe())
```

```python
# quick summary for each class
df.groupby('class').mean()
```

```python
plt.scatter(df['sepal_length'], df['sepal_width'], label='class')
```

```python
# Visulize the summary

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, dpi=200)

# plot sepal length v width

ax1.set_title('Sepal Sizes (cm)')
for cls in classes:
    d = df[df['class'] == cls]
    ax1.scatter(d['sepal_length'], d['sepal_width'], label=cls)

ax1.set_xlabel('length (cm)')
ax1.set_ylabel('width (cm)')
ax1.legend()

# plot petal length v width

ax2.set_title('Petal Sizes (cm)')
for cls in classes:
    d = df[df['class'] == cls]
    ax2.scatter(d['petal_length'], d['petal_width'], label=cls)
    
ax2.set_xlabel('length (cm)')
ax2.set_ylabel('width (cm)')
ax2.legend()
```

What does it tell?

* setosa has distinctive petal and sepal sizes
* although versicolor and virginica have similar sepal sizes, petal sizes can be used to dintinguish the two

Classification problem - use KNN

* telling the difference between versicolor and virginica would be more difficult compared to isolating setosa

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df.drop('class', axis=1)
y = df['class']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)
```

```python
y_predict = knn_model.predict(X_test)
```

```python
knn_model.score(X_test, y_test)
```
